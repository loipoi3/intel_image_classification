```python
# import packages
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
def count_files(path):
    counts = {}
    for root, dirs, files in os.walk(path):
        for file in files:
            class_name = os.path.basename(root)
            counts[class_name] = counts.get(class_name, 0) + 1
    return counts
```

```python
train_path = "./data/seg_train"
test_path = "./data/seg_test"
train_counts = count_files(train_path)
test_counts = count_files(test_path)
```

```python
classes = train_counts.keys()
x_pos = np.arange(len(classes))
train_values = train_counts.values()
test_values = test_counts.values()
```

```python
plt.bar(x_pos, train_values, align='center', alpha=0.5, label='Train')
plt.xticks(x_pos, classes, rotation='vertical')
plt.ylabel('Amount of elements')
plt.xlabel('Classes')
plt.title('Histogram of the number of elements by class')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.bar(x_pos, test_values, align='edge', alpha=0.5, label='Test')
plt.xticks(x_pos, classes, rotation='vertical')
plt.ylabel('Amount of elements')
plt.xlabel('Classes')
plt.title('Histogram of the number of elements by class')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
print(f'The number of images for each class in the train folder: {train_counts}')
print(f'The number of images for each class in the test folder: {test_counts}')
```

```python
def check_umbalanced_classes(dict_cls):
    max_class = max(dict_cls.values())
    min_class = min(dict_cls.values())

    max_percentage = (max_class / sum(dict_cls.values())) * 100
    min_percentage = (min_class / sum(dict_cls.values())) * 100

    deviation = max_percentage - min_percentage

    threshold = 5

    if deviation > threshold:
        print("Classes are unbalanced.")
    else:
        print("Classes are balanced.")
```

```python
check_umbalanced_classes(train_counts)
```

```python
check_umbalanced_classes(test_counts)
```
